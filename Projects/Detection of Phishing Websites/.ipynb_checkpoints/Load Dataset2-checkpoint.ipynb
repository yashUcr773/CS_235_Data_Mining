### Import the dataset

```python
path_to_dataset_csv = './Phishing_Legitimate_full.csv'
data = pd.read_csv(path_to_dataset_csv,index_col=0)
data.shape
```

### Drop any null/nan values

```python
data = data.dropna()
data.shape
```

### Check for counts of different values

```python
cols = list(data.columns)
for col in cols:
    print(col)
    cnt = data[col].value_counts()
    print (cnt.shape)
    
    if cnt.shape[0] == 1:
        print (col , ' only has 1 attribute so this can be removed as it wont affect our prediction')
        
```

```python
data = data.drop(index=('HttpsInHostname'), axis = 1)
```

### Check for co-relation between different features

```python
plt.figure(figsize=(30, 20))
sns.heatmap(data.corr().abs(),  annot=True)
```

#### Using co-relation we can check which features affect the result more than others.

```python
mean_of_corr_for_result = sum(data.corr().abs()['CLASS_LABEL'][:-1])/len(data.corr().abs()['CLASS_LABEL'][:-1])
cols_with_corr_greater_than_mean = {}
print (mean_of_corr_for_result)

for col in data.corr().abs()['CLASS_LABEL'][:-1].index:
    print (data.corr().abs()['CLASS_LABEL'][:-1][col])
    if data.corr().abs()['CLASS_LABEL'][:-1][col] > mean_of_corr_for_result:
        cols_with_corr_greater_than_mean[col] = data.corr().abs()['CLASS_LABEL'][:-1][col]
print (cols_with_corr_greater_than_mean)
```

```python
data2 = data[cols_with_corr_greater_than_mean.keys()]
```

```python
data2
```

```python
X = data2
Y = data['Result']
train_data, test_data, train_labels, test_labels = train_test_split(X, Y, test_size=0.3)
print (train_data.shape, train_labels.shape)
print (test_data.shape, test_labels.shape)
```

```python
from sklearn.linear_model import Perceptron
from sklearn.metrics import accuracy_score

test_loss = []
train_loss = []
p = Perceptron(tol=None, max_iter = 1,n_iter_no_change=100,alpha=0.0000001,warm_start=True)

for epoch in range(100):
    
    print('epoch ->', epoch)
    p.fit(train_data, train_labels)

    predictions_train = p.predict(train_data)
    predictions_test = p.predict(test_data)

    train_score = accuracy_score(predictions_train, train_labels)
    test_score = accuracy_score(predictions_test, test_labels)

    test_loss.append(test_score)    
    train_loss.append(train_score)

plt.plot(test_loss[:10])
plt.plot(train_loss[:10])
```

### T-SNE for data

```python
from sklearn.manifold import TSNE
 
# We want to get TSNE embedding with 2 dimensions
tsne = TSNE(2)
tsne_result = tsne.fit_transform(train_data)
tsne_result.shape
```

```python
tsne_result
```

```python
phishing_result = tsne_result[train_labels == -1]
benign_result = tsne_result[train_labels == 1]

print(phishing_result.shape)
print(benign_result.shape)
```

```python
plt.figure(figsize=(20,20))
plt.scatter(phishing_result[:,0], phishing_result[:,1], color='red',s=15)
plt.scatter(benign_result[:,0], benign_result[:,1], color='green',s=5)
plt.show()
```
